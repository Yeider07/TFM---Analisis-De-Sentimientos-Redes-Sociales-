# Librerias

```python
import json
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from deep_translator import GoogleTranslator
import pandas as pd
```

# Cargar Archivo

```python
# Nombre del archivo JSON
archivo_entrada = "DB Completo Sin Sentimientos.json"

# Cargar el archivo JSON
try:
    with open(archivo_entrada, "r", encoding="utf-8") as json_file:
        publicaciones = json.load(json_file)
except Exception as e:
    print(f"Error al cargar el archivo JSON: {e}")
    exit()

# Convertir a DataFrame
data = pd.json_normalize(
    publicaciones,
    record_path='comentarios',
    meta=['titulo', 'pagina', 'fecha', 'reacciones', 'veces_compartido', 'numero_comentarios', 'hashtags'],
    errors='ignore'
)
```

```python
print(type(data))  # Esto te dirá qué tipo de objeto es 'data'
print(data)        # Esto te mostrará el contenido de 'data'
```

# Traducir a Ingles comentarios 

## Procesar los Comentarios


```python
# Procesar los comentarios
for publicacion in data:
    titulo = publicacion.get('titulo', '')
    fecha = publicacion.get('fecha', '')
    for comentario in publicacion.get('comentarios', []):
        usuario = comentario.get('usuario', '')
        texto_comentario = comentario.get('comentario', '')

        # Verificar que el comentario no sea None o vacío
        if not texto_comentario:
            texto_comentario_traducido = ""
        else:
            # Traducir el comentario al inglés
            try:
                texto_comentario_traducido = GoogleTranslator(source='auto', target='en').translate(texto_comentario)
            except Exception as e:
                print(f"Error al traducir el comentario: {texto_comentario}. Error: {e}")
                texto_comentario_traducido = texto_comentario

        # Almacenar los datos
        comentarios_data.append({
            'titulo': titulo,
            'fecha': fecha,
            'usuario': usuario,
            'comentario': texto_comentario,
            'comentario_traducido': texto_comentario_traducido
        })

# Crear un DataFrame con los datos traducidos
df_traducido = pd.DataFrame(comentarios_data)

# Guardar los resultados en un archivo JSON
output_file = 'comentarios_traducidos Data Completa.json'
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(comentarios_data, json_file, ensure_ascii=False, indent=4)

print(f"Traducción completada. Resultados guardados en '{output_file}'.")
```

```python
# Procesar los comentarios directamente desde el DataFrame
comentarios_data = []

for _, row in data.iterrows():  # Iterar sobre las filas del DataFrame
    titulo = row['titulo']
    fecha = row['fecha']
    usuario = row['usuario'] if 'usuario' in row else ''
    texto_comentario = row['comentario'] if 'comentario' in row else ''

    # Verificar que el comentario no sea None o vacío
    if not texto_comentario:
        texto_comentario_traducido = ""
    else:
        # Traducir el comentario al inglés
        try:
            texto_comentario_traducido = GoogleTranslator(source='auto', target='en').translate(texto_comentario)
        except Exception as e:
            print(f"Error al traducir el comentario: {texto_comentario}. Error: {e}")
            texto_comentario_traducido = texto_comentario

    # Almacenar los datos
    comentarios_data.append({
        'titulo': titulo,
        'fecha': fecha,
        'usuario': usuario,
        'comentario': texto_comentario,
        'comentario_traducido': texto_comentario_traducido
    })

# Crear un DataFrame con los datos traducidos
df_traducido = pd.DataFrame(comentarios_data)

# Guardar los resultados en un archivo JSON
output_file = 'DB Traducido.json'
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(comentarios_data, json_file, ensure_ascii=False, indent=4)

print(f"Traducción completada. Resultados guardados en '{output_file}'.")
```

# Analizar sentimiento con TextBlob

```python
# Crear una lista para almacenar los resultados
sentimientos_data = []

# Analizar sentimientos para cada comentario traducido
def analizar_sentimientos(df):
    for index, row in df.iterrows():
        comentario = row['comentario']
        comentario_traducido = row['comentario_traducido']

        # Verificar que el comentario traducido no esté vacío
        if not comentario_traducido or not isinstance(comentario_traducido, str):
            polaridad = 0.0
            sentimiento = "Neutral"
        else:
            # Analizar sentimiento con TextBlob
            blob = TextBlob(comentario_traducido)
            polaridad = blob.sentiment.polarity  # Valor entre -1 (negativo) y 1 (positivo)

            # Asignar categorias a las polaridad para mayor entendimiento
            if polaridad > 0:
                sentimiento = "Positivo"
            elif polaridad < 0:
                sentimiento = "Negativo"
            else:
                sentimiento = "Neutral"

        # Agregar los resultados a la lista
        sentimientos_data.append({
            'comentario': comentario,
            'comentario_traducido': comentario_traducido,
            'polaridad': polaridad,
            'sentimiento': sentimiento
        })

    # Convertir la lista en un DataFrame
    return pd.DataFrame(sentimientos_data)

# Llamar a la función con df_traducido y obtener los resultados
df_sentimientos = analizar_sentimientos(df_traducido)

# Mostrar los primeros resultados
print(df_sentimientos.head())

# Guardar los resultados en un archivo JSON
output_file = 'Sentimientos_Traducidos.json'
df_sentimientos.to_json(output_file, orient='records', force_ascii=False, indent=4)
print(f"Análisis de sentimientos completado. Resultados guardados en '{output_file}'.")
```

```python
# Graficar los resultados
conteo_sentimientos = Counter(df_sentimientos['sentimiento'])

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(conteo_sentimientos.keys(), conteo_sentimientos.values(), alpha=0.7, color=['green', 'red', 'blue'])
plt.title('Distribución de Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

# Análisis de palabras más comunes por sentimiento
def palabras_mas_comunes_por_sentimiento(df, sentimiento):
    comentarios = " ".join(df[df['sentimiento'] == sentimiento]['comentario_traducido'].dropna())
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(comentarios)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Palabras más comunes - {sentimiento}', fontsize=16)
    plt.axis('off')
    plt.show()

# Generar nubes de palabras para cada sentimiento
for sentimiento in ['Positivo', 'Negativo', 'Neutral']:
    palabras_mas_comunes_por_sentimiento(df_sentimientos, sentimiento)
```

```python
# Obtener el top 30 de palabras más utilizadas por sentimiento
def top_palabras_por_sentimiento(df, sentimiento, top_n=30):
    comentarios = " ".join(df[df['sentimiento'] == sentimiento]['comentario_traducido'].dropna())
    palabras = comentarios.split()
    conteo_palabras = Counter(palabras)
    top_palabras = conteo_palabras.most_common(top_n)

    print(f"\nTop {top_n} palabras más comunes para el sentimiento '{sentimiento}':")
    for palabra, frecuencia in top_palabras:
        print(f"{palabra}: {frecuencia}")

# Mostrar el top 30 de palabras por cada sentimiento
for sentimiento in ['Positivo', 'Negativo', 'Neutral']:
    top_palabras_por_sentimiento(df_sentimientos, sentimiento)
```

```python
# Obtener las palabras con mayor peso por sentimiento
def palabras_mayor_peso_por_sentimiento(df, sentimiento, top_n=100):
    palabras_polaridad = []

    # Filtrar por sentimiento
    comentarios = df[df['sentimiento'] == sentimiento]

    for comentario in comentarios['comentario_traducido'].dropna():
        blob = TextBlob(comentario)
        for palabra in blob.words:
            polaridad = TextBlob(palabra).sentiment.polarity
            palabras_polaridad.append((palabra, polaridad))

    # Ordenar por polaridad y obtener las top_n palabras
    palabras_polaridad.sort(key=lambda x: x[1], reverse=(sentimiento == 'Positivo'))
    top_palabras = palabras_polaridad[:top_n]

    print(f"\nTop {top_n} palabras con mayor peso para el sentimiento '{sentimiento}':")
    for palabra, polaridad in top_palabras:
        print(f"{palabra}: {polaridad:.2f}")

# Mostrar el top 10 de palabras con mayor peso por cada sentimiento
for sentimiento in ['Positivo', 'Negativo', 'Neutral']:
    palabras_mayor_peso_por_sentimiento(df_sentimientos, sentimiento)
```

```python
# Obtener las palabras con mayor peso por sentimiento y guardarlas en un Excel
def palabras_mayor_peso_por_sentimiento(df, sentimiento, top_n=100000):
    palabras_polaridad = []

    # Filtrar por sentimiento
    comentarios = df[df['sentimiento'] == sentimiento]

    for comentario in comentarios['comentario_traducido'].dropna():
        blob = TextBlob(comentario)
        for palabra in blob.words:
            polaridad = TextBlob(palabra).sentiment.polarity
            palabras_polaridad.append({'palabra': palabra, 'polaridad': polaridad, 'sentimiento': sentimiento})

    # Ordenar por polaridad y obtener las top_n palabras
    palabras_polaridad = sorted(palabras_polaridad, key=lambda x: x['polaridad'], reverse=(sentimiento == 'Positivo'))[:top_n]
    return palabras_polaridad

# Crear un DataFrame para almacenar todas las palabras con sus pesos y emociones
todas_palabras = []
for sentimiento in ['Positivo', 'Negativo', 'Neutral']:
    todas_palabras.extend(palabras_mayor_peso_por_sentimiento(df_sentimientos, sentimiento))

# Convertir a DataFrame y guardar en un archivo Excel
df_palabras_pesos = pd.DataFrame(todas_palabras)
output_excel_file = 'Palabras_Pesos_Sentimientos.xlsx'
df_palabras_pesos.to_excel(output_excel_file, index=False)
print(f"Archivo Excel generado: {output_excel_file}")
```
